```python
import re

# 读取 log_xx.txt 文件
with open(file_path_2, 'r', encoding='utf-8') as f:
    lines = f.readlines()

train_process_2 = []  # 用于保存提取的 Train-Epoch、Loss、Acc 数组

# 遍历每一行
for line in lines:
    # 使用正则表达式匹配 Train-Epoch、Loss、Acc 数字
    match = re.search(r'Train-Epoch:\s*(\d+).*Loss:\s*([\d.]+).*Acc:\s*([\d.]+)', line)
    if match:
        # 提取 Train-Epoch、Loss、Acc 数字并转换为 float 类型
        train_epoch = int(match.group(1))
        # 将 "NaN" 转换为 -1
        loss = float(match.group(2)) if match.group(2) != "NaN" else -1
        acc = float(match.group(3)) if match.group(3) != "NaN" else -1
        # 将 Train-Epoch、Loss、Acc 数字放入 tmp 数组中
        tmp = [train_epoch, loss, acc]
        # 将 tmp 数组放入 train_process 数组中
        train_process_2.append(tmp)

print(train_process_2)
```

```python
import re

# 读取 log_xx.txt 文件
with open(file_path, 'r', encoding='utf-8') as f:
    lines = f.readlines()

train_process = []  # 用于保存提取的 Train-Epoch、Loss、Acc 数组

# 遍历每一行
for line in lines:
    # 使用正则表达式匹配 Train-Epoch、Loss、Acc 数字
    match = re.search(r'Train-Epoch:\s*(\d+).*Loss:\s*([\d.]+).*Acc:\s*([\d.]+)', line)
    if match:
        # 提取 Train-Epoch、Loss、Acc 数字并转换为 float 类型
        train_epoch = int(match.group(1))
        # 将 "NaN" 转换为 -1
        loss = float(match.group(2)) if match.group(2) != "NaN" else -1
        acc = float(match.group(3)) if match.group(3) != "NaN" else -1
        # 将 Train-Epoch、Loss、Acc 数字放入 tmp 数组中
        tmp = [train_epoch, loss, acc]
        # 将 tmp 数组放入 train_process 数组中
        train_process.append(tmp)

print(train_process)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# 提取 Train-Epoch、Loss、Acc 数据
train_epochs = [item[0] for item in train_process]
losses = [item[1] for item in train_process]
accs = [item[2] for item in train_process]

# 设置 Seaborn 绘图风格
sns.set_style('whitegrid')  # 设置白色网格风格
sns.set_palette('husl')  # 设置颜色调色板

# 创建画布和坐标轴
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.set_xlabel('Train-Epoch')  # 设置 x 轴标签
ax1.set_ylabel('Loss', color='tab:blue')  # 设置 y 轴标签（Loss），设置颜色为蓝色
sns.lineplot( losses, label='Loss', ax=ax1, color='tab:blue')  # 绘制 Loss 曲线
ax1.tick_params(axis='y', labelcolor='tab:blue')  # 设置 y 轴刻度标签颜色为蓝色
ax1.set_ylim(-1,max(losses))

# 创建第二个坐标轴
ax2 = ax1.twinx()
ax2.set_ylabel('Acc', color='tab:red')  # 设置 y 轴标签（Acc），设置颜色为红色
sns.lineplot(accs, label='Acc', ax=ax2, color='tab:red')  # 绘制 Acc 曲线
ax2.tick_params(axis='y', labelcolor='tab:red')  # 设置 y 轴刻度标签颜色为红色
# 设置 Acc 坐标轴刻度范围
ax2.set_ylim(0, 1)

ax1.grid(True)
ax2.grid(False)

plt.title('Loss and Acc over Train-Epoch')  # 设置图标题
plt.legend()  # 添加图例
plt.show()  # 显示图像
```
